```python
import os
import shutil
import random
from tqdm import tqdm
```

```python
Dataset_root = 'SJB_25_Dataset'
```

## åˆ’åˆ†è®­ç»ƒé›†ã€æµ‹è¯•é›†

```python
test_frac = 0.2  # æµ‹è¯•é›†æ¯”ä¾‹
random.seed(123) # éšæœºæ•°ç§å­ï¼Œä¾¿äºå¤ç°
```

```python
folder = '.'
```

```python
img_paths = os.listdir(folder)
random.shuffle(img_paths) # éšæœºæ‰“ä¹±

val_number = int(len(img_paths) * test_frac) # æµ‹è¯•é›†æ–‡ä»¶ä¸ªæ•°
train_files = img_paths[val_number:]         # è®­ç»ƒé›†æ–‡ä»¶ååˆ—è¡¨
val_files = img_paths[:val_number]           # æµ‹è¯•é›†æ–‡ä»¶ååˆ—è¡¨

print('æ•°æ®é›†æ–‡ä»¶æ€»æ•°', len(img_paths))
print('è®­ç»ƒé›†æ–‡ä»¶ä¸ªæ•°', len(train_files))
print('æµ‹è¯•é›†æ–‡ä»¶ä¸ªæ•°', len(val_files))
```

## å°†è®­ç»ƒé›†å›¾åƒç§»åŠ¨åˆ°`images/trian`ç›®å½•

```python
os.mkdir('train')
for each in tqdm(train_files):
    shutil.move(each, 'train')
```

## å°†æµ‹è¯•é›†å›¾åƒç§»åŠ¨åˆ°`images/val`ç›®å½•

```python
os.mkdir('val')
for each in tqdm(val_files):
    shutil.move(each, 'val')
```

## å°†æµ‹è¯•é›†æ ‡æ³¨ç§»åŠ¨åˆ°`labelme_jsons/val`ç›®å½•

```python
os.mkdir('val')
for each in tqdm(val_files):
    srt_path = each.split('.')[0] + '.json'
    shutil.move(srt_path, 'val')
```
